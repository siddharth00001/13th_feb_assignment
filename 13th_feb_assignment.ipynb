## Q1

We inherit exception to get the behaviour and attributes of base exception class and we can then override this behavior to work according to our needs.

```python
class custEx(Exception):
    
    def __init__(self,msg):
        self.msg = msg

def TypeCheck(var):
    if var == str:
        print("format is correct")
    else:
        raise custEx("format is incorrect")

try:
    data = int(input())
    TypeCheck(data)
except custEx as e:
    print(e)
```

## Q3

The errors defined inside arithmetics error deals with mainly mathematical operation like division by zero error or Decimal exception error.
1) ZeroDivisionError error is the error which gets raise when we try to divide an integer with zero.
2) An OverflowError exception is raised when an arithmetic operation exceeds the limits to be represented. This is part of the ArithmeticError Exception class.

```python
j=5.0
try:
    for i in range(1000):
        print(j**i)
except OverflowError as e:
    print(e)
```

## Q4

The lookup error is the base class for all exceptions that arise when the index or key is not found in a sequence or dictionary. The two types of error that comes under the lookup error is keyerror and index error.

```python
a='str'
try:
    for x in range(5):
        print(a[x])
except (KeyError,IndexError) as e:
    print(f'This error {e} happend')
```

```python
k={1:'a',2:'b'}
try:
    print(k['r'])
except KeyError as e:
    print(f'This error {e} happend')
```

## Q5

Import error happens mainly of the three reasons listed below:-
1) When there is circular dependency that is when we import 2 different modules in each other.
2) when the name of module is mispelt or it is incorrect.
3) The imported module is not available in python.

Module not found error happens usually when the module we are trying to import is not installed. If we try to import additional module other than module present by default we need to install it using pip(pip install packages).

## Q6

Some best practices are listed below :-
1) Always try to run the code in try and except command. so that if code fails it doesn't look ugly and we can debug it easily.
2) Always try to be specific about the error to catch if we get an error.
3) Always add single exception line to catch instead of adding multiple except conditions for catching errors.
4) Instead of printing the error try to log the error so it can be debuged later if it fails in production.

```python
import logging
```

```python
logging.basicConfig(filename='eror.log',level=logging.DEBUG,format='%(asctime)s %(name)s %(message)s')
try:
    
    print(int('hello'))
    logging.debug("Code Worked")
except ValueError as e:
    logging.error(f'following exception {e} happend')
logging.shutdown()
```
